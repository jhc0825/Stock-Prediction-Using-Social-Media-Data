```python
stock_quotes = pd.DataFrame()
for c in companies:
    r = requests.get('https://finnhub.io/api/v1/stock/candle?symbol=' + c + '&resolution=D&from=' + start+ '&to=' + end + '&token=bu6l53n48v6pfj0oklv0')
    y = json.loads(json.dumps(r.json()))
    t = [""]*len(y["t"])
    for x in range (0,len(t)):
        t[x] = datetime.datetime.fromtimestamp(y["t"][x]).strftime('%y-%m-%d')
    company_quote = pd.DataFrame({
    'Company' : [c]*len(y["v"]),
    'Open' : np.round(y["o"]),
    'Close' : np.round(y["c"]),
    'Volume' : y["v"],
    'Date' : t
    })
    stock_quotes = pd.concat([stock_quotes, company_quote], axis = 0).reset_index(drop=True)
```
